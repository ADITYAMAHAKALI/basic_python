### Function

input -> process -> output
e.g. print the table of a number

input : number n
process : multiply n with 1 to 10
output : print the table, None

```python
# function definition
def printTable(n:int) -> None:
    """
    Print the multiplication table of a given number n from 1 to 10.
    Args:
        n (int): The number for which the multiplication table is to be printed.
    Returns:
        None"""
    for i in range(1, 11):
        print(f"{n} x {i} = {n*i}")
```

```python
# function call
printTable(7)
```

```python
printTable(6)
```

## Search function

```python
l = [6, 12, 18, 24, 30, 36, 42, 48, 54, 60]

# to get the index of an element
x = 48

for idx, element in enumerate(l):
    if element == x:
        print(f"Index of {x} is {idx}")
        break
```

```python
def searchElement(l: list[int], x:int) -> int:
    """
    Search for an element x in the list l and return its index.
    If the element is not found, return -1.
    Args:
        l (list[int]): The list in which to search for the element.
        x (int): The element to search for.
    Returns:
        int: The index of the element if found, otherwise -1.
    """
    for idx, element in enumerate(l):
        if element == x:
            return idx
    return -1
```

```python
searchElement(l, 48)
```

```python
def binSearch(l : list[int], x: int)-> int:
    lo = 0 
    hi = len(l) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if l[mid] == x:
            return mid
        elif l[mid] < x:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1
```

```python
binSearch(l, 48)
```

```python
class point:
    x: float
    y: float
    # constructor
    def __init__(self, x: float, y: float) -> None:
        self.x = x
        self.y = y
    
    def display(self) -> None:
        print(f"Point({self.x}, {self.y})")
    
    def distance(self,other: 'point') -> float:
        return ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
```

```python
# class definition
class Car:
    """
    A class to represent a car with its name, model, and maximum speed.
    Attributes:
        name (str): The name of the car.
        model (str): The model of the car.
        max_speed (float): The maximum speed of the car in km/h."""
    name: str
    model: str
    max_speed: float
    
    def __init__(self, name: str, model: str, max_speed: float) -> None:
        self.name = name
        self.model = model
        self.max_speed = max_speed
    
    def display(self) -> None:
        """
        Display the car's details.
        Returns:
            None"""
        print(f"Car Name: {self.name}, Model: {self.model}, Max Speed: {self.max_speed} km/h")
```

```python
class Account:
    account_number : str
    balance : float
    account_owner_name : str
    
    def __init__(self, account_number: str, balance: float, account_owner_name: str) -> None:
        self.account_number = account_number
        self.balance = balance
        self.account_owner_name = account_owner_name
    
    def checkBalance(self) -> float:
        return self.balance
    
    def deposit(self, amount: float) -> str:
        self.balance += amount
        return "Deposit successful, new balance is " + str(self.balance)
    
    def withdraw(self, amount: float) -> str:
        if amount > self.balance:
            return "Insufficient balance"
        else:
            self.balance -= amount
            return "Withdrawal successful, new balance is " + str(self.balance)
    
    def transfer(self, amount: float, to_account: 'Account') -> str:
        if amount > self.balance:
            return "Insufficient balance"
        else:
            self.balance -= amount
            to_account.balance += amount
            return "Transfer successful, new balance is " + str(self.balance)
    
```

```python
acc = Account(account_number="1234567890", balance=1000.0, account_owner_name="John Doe")
```

```python
acc.checkBalance()
```

```python
acc.deposit(500.0)
```

```python
acc.withdraw(2000.0)
```

```python
acc.checkBalance()
```

```python
acc2 = Account(account_number="0987654321", balance=500.0, account_owner_name="Jane Smith")
```

```python
acc2.checkBalance()
```

### File Handling in Python

- read from file : 'r'
- write to file : 'w'
- append to file : 'a'

```python
content = ""
with open("./files/demo.txt", "r") as f:
    content = f.read()
```

```python
len(content)
```

- .  means current directory
- .. means parent directory
- ... means grandparent directory

```python
content += "\n"+ str(len(content))
with open("./files/demo.txt", "w") as f:
    f.write(content)
```

```python
def chunking_file(file_path: str, chunk_size: int) -> list[str]:
    chunks = []
    with open(file_path, "r") as f:
        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break
            chunks.append(chunk)
    return chunks
```

```python
chunking_file("./files/demo.txt", 100)
```
