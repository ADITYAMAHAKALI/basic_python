```python
# whether the person can drive or not
age = int(input("Enter your age"))
if age >=18:
   print("Yes")
else:
    print("No")  
```

```python
# whether the person can drive and marry or not
age = int(input("Enter your age"))
if age >=18:
   print("Can drive")
   if age >=21:
        print("Can marry")
   else:
        print("Can not marry")
else:
    print("Can not drive and can not marry")  
```

```python
# whether the person can drive and marry or not
age = int(input("Enter your age"))
if age >=21:
   print("Can drive and marry")
elif age >=18 and age <21:
    print("Can drive but can not marry")
else :
    print("Can not drive and can not marry") 
```

## Loops

- for loop
- while loop

```python
for i in range(0,21): # end is exclusive
    print(i)
```

```python
# print a table of a number
num = int(input("Enter a number"))
# num * 1, num *2 ... num *10
for mul in range(1,11):
    print(f"{num} x {mul} = {num * mul}")
```

```python
# print even number upto n
num = int(input("Enter a number"))
# num * 1, num *2 ... num *10
for i in range(0,num+1,2):
    print(f"{i}")
```

```python
# reverse order
for num in range(10,0,-1):
    print(num)
```

```python
l = [6, 12, 18, 24, 30, 36, 42, 48, 54, 60]
# calculate the sum of elements in l
total = 0
for element in l:
    total += element
total
```

```python
l = [6, 12, 18, 24, 30, 36, 42, 48, 54, 60]
# calculate the sum of all even index elements in l
total = 0
idx = 0
for element in l:
    print(idx)
    if idx % 2 == 0:
        total += element
    idx += 1
total
```

```python
l = [6, 12, 18, 24, 30, 36, 42, 48, 54, 60]
# calculate the sum of all even index elements in l
total = 0
for idx,element in enumerate(l):
    if idx % 2 == 0: # whether the index is even
        print(idx,element)
        total += element
    idx += 1
total
```
